```python
tokens = spark.read.load('gs://project_joyce/combined_tokens').sample(False, 0.01).select(['tokens']).take(100000)
```

```python
def fix_unicode(text):
  unicode_dict = {
    u'\u2019': '\'',
    u'\u2014': '--',
    u'\u201d': '"',
    u'\u201c': '"',
    u'\u2026': '...',
    u'\u2018': '\'',
    u'\u2013': '-',
    u'\x91': '\'',
    u'\x92': '\'',
    u'\u2011': '-',
    '\'\'': '"',
    '``': '"',
  }
  for uni, replacement in unicode_dict.items():
    text = text.replace(uni, replacement)
  return text

for token_set in tokens:
  try:
    fix_unicode(' '.join(token_set.tokens)).decode('utf-8', 'ignore')
  except UnicodeEncodeError as e:
    print(e)
```
