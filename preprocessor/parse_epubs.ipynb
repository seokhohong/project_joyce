```python
rootdir = '/home/jupyter/top_fiction'
epub_dir = 'epub_files'

for root, dirs, files in os.walk(rootdir):
    for file in files:
        print(file)
        if file.endswith('.epub') and epub_dir not in root:
            input_filename = os.path.join(os.getcwd(), root, file)
            output_filename = os.path.join(os.getcwd(), rootdir, epub_dir, re.sub('[^0-9a-zA-Z.]+', '_', file))
            print(output_filename)
            try:
                result = subprocess.check_output(['cp', 
                                                  input_filename,
                                                  output_filename])
            except subprocess.CalledProcessError as e:
                print(e)
```
